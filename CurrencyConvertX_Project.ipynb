```python
import requests
import json

def fetch_exchange_rates(url):
    response = requests.get(url)
    if not response.ok:
        print("\nError {}: ".format(response.status_code))
        error_message = response.json().get('error', 'Unknown error')
        print(error_message)
        return None
    else:
        return response.json()

# Input from user
base_currency = input("\nConversion from ('currency'): ")

base_url_template = "https://v6.exchangerate-api.com/v6/961e0c53f479988e0cbe2ae1/latest/{}"
base_url = base_url_template.format(base_currency)

exchange_rates = fetch_exchange_rates(base_url)

print("Hello!\nWelcome to CurrencyConvertX, a brand new Currency Exchanger Software\nUseful for conversion of any currency rates.. \t\t\t ANY!\n")
print("\nGiven below are the list of Currency rates you can choose for Conversion:")

if exchange_rates:
    print(json.dumps(exchange_rates['conversion_rates'], indent=4))
    target_currency = input("\nConversion to ('currency'): ")
    quan = float(input("\nHow much {} do you want to convert: ".format(base_currency)))

    def convert_currency(exchange_rates, target_currency, quan):
        if target_currency in exchange_rates['conversion_rates']:
            rate = exchange_rates['conversion_rates'][target_currency]
            result = rate * quan
            return result
        else:
            print("\nError: Target currency not found in exchange rates.")
            return None

    result = convert_currency(exchange_rates, target_currency, quan)
    if result is not None:
        print("\n'{0} {1}' is equal to '{2:.2f} {3}' based upon the latest exchange rates in 2024!\n\nThank you for using CurrencyConvertX! Have a good day ^-^.".format(quan, base_currency, result, target_currency))
    else:
        print("\nError: Could not convert currency. Please try again later.")
else:
    print("\nError: Could not fetch exchange rates. Please try again later.")
```
